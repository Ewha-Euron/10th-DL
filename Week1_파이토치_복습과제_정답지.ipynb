## **문제 1: 텐서(Tensor) 기초 조작**
### **1-1. 텐서 생성 및 차원 변경 & 1-2. 디바이스 할당**

```python
import torch

# 1-1. 1부터 6까지의 1차원 텐서 생성 및 2x3 2차원 텐서로 변환
temp_tensor = torch.tensor([1, 2, 3, 4, 5, 6])
reshaped_tensor = temp_tensor.view(2, 3) # reshape(2, 3)도 정답으로 인정합니다.
print("차원 변경 후 텐서:\n", reshaped_tensor)

# 1-2. 디바이스 할당
if torch.cuda.is_available():
    device = torch.device('cuda')
else:
    device = torch.device('cpu')

tensor_on_device = reshaped_tensor.to(device)
print(f"\n할당된 디바이스: {tensor_on_device.device}")
```

## **문제 2: 커스텀 데이터셋 구축**

```python
from torch.utils.data import Dataset
import pandas as pd

class MyCustomDataset(Dataset):
    # 1. 초기화 메서드
    def __init__(self, csv_file):
        self.data = pd.read_csv(csv_file)

    # 2. 데이터 길이 반환 메서드
    def __len__(self):
        return len(self.data)

    # 3. 특정 인덱스 데이터 반환 메서드
    def __getitem__(self, idx):
        sample = torch.tensor(self.data.iloc[idx, 0:-1].values).float()
        label = torch.tensor(self.data.iloc[idx, -1]).long()
        return sample, label

print("MyCustomDataset 클래스 선언이 완료되었습니다.")
```

## **문제 3: Sequential을 이용한 모델 정의**

```python
import torch.nn as nn

class SimpleMLP(nn.Module):
    def __init__(self, input_size, output_size):
        super(SimpleMLP, self).__init__()
        # nn.Sequential을 이용한 네트워크 구조 정의
        self.network = nn.Sequential(
            nn.Linear(in_features=input_size, out_features=20),
            nn.ReLU(inplace=True),
            nn.Linear(in_features=20, out_features=output_size)
        )

    def forward(self, x):
        return self.network(x)

model = SimpleMLP(input_size=10, output_size=2)
print("생성된 모델 구조:\n", model)
```

## **문제 4: 모델 훈련 루프 5단계**

```python
import torch
import torch.nn as nn

# 가상의 모델, 오차 함수, 옵티마이저 설정 (코드가 실행되도록 임의 구성)
model = SimpleMLP(10, 2)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# 가상의 1개 배치 데이터 생성 (배치 크기 4, 입력 특징 10) 및 정답(레이블) 설정
x_batch = torch.randn(4, 10)
y_batch = torch.tensor([0, 1, 0, 1])

print("=== 훈련 루프 1회 진행 ===")

# 1. 기울기 초기화
optimizer.zero_grad()

# 2. 모델 예측(출력 계산)
y_pred = model(x_batch)

# 3. 오차 계산
loss = criterion(y_pred, y_batch)
print(f"현재 배치 오차(Loss): {loss.item():.4f}")

# 4. 역전파 학습 기울기 자동 계산
loss.backward()

# 5. 기울기 업데이트
optimizer.step()

print("=== 훈련 루프 완료 ===")
```
